5. The following is a deÔ¨Ånition of the class Point that takes in the coordinates x, y. Add a method
plot_point that plots the position of a point.

```python
import matplotlib.pyplot as plt

class Point():
    def __init__(self, x, y):
        self.x = x 
        self.y = y
    def plot_point(self):
        plt.scatter(self.x, self.y)
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.show()
```

```python
p = Point(1,2)
p.plot_point()
```

6. Use the class from Problem 5 and add a method calculate_dist which takes in x and y from
another point, and returns the distance calculated between the two points.

```python
import matplotlib.pyplot as plt

class Point():
    def __init__(self, x, y):
        self.x = x 
        self.y = y
    def plot_point(self):
        plt.scatter(self.x, self.y)
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.show()
    def calculate_dist(self, q):
        return ((self.x-q.x)**2 + (self.y-q.y)**2)**0.5

p = Point(1,2)
q = Point(4,3)
print(f"Distance between p and q is {p.calculate_dist(q)}")
```
